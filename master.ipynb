```python
# Import libraries
import pandas as pd
from pathlib import Path
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
#Load CSVs
file_path_2019 = Path("RESOURCES/2019.csv")
file_path_2018 = Path("RESOURCES/2018.csv")
file_path_2017 = Path("RESOURCES/2017.csv")
file_path_2016 = Path("RESOURCES/2016.csv")
file_path_2015 = Path("RESOURCES/2015.csv")
file_path_2014 = Path("RESOURCES/2014.csv")
file_path_2013 = Path("RESOURCES/2013.csv")
file_path_2012 = Path("RESOURCES/2012.csv")
file_path_2011 = Path("RESOURCES/2011.csv")
#Read in CSVs
df_2019 = pd.read_csv(file_path_2019)
df_2018 = pd.read_csv(file_path_2018)
df_2017 = pd.read_csv(file_path_2017)
df_2016 = pd.read_csv(file_path_2016)
df_2015 = pd.read_csv(file_path_2015)
df_2014 = pd.read_csv(file_path_2014)
df_2013 = pd.read_csv(file_path_2013)
df_2012 = pd.read_csv(file_path_2012)
df_2011 = pd.read_csv(file_path_2011)
```

```python
#Concat Data
combined_epl_df = pd.concat([df_2019, df_2018, df_2017, df_2016, df_2015, df_2014, df_2013, df_2012, df_2011],axis="index", join="inner")
#Dataframe converted from British Date to US Date
combined_epl_df['Date'] = pd.to_datetime(combined_epl_df["Date"])
combined_epl_df["Date"] = combined_epl_df["Date"].dt.strftime("%m-%d-%Y")
combined_epl_df.head()
```

```python
#Set Date as index
combined_epl_df.set_index('Date', inplace = True)
combined_epl_df.head()
```

```python
# Check data for nulls, check for sum...
combined_epl_df.dropna()
combined_epl_df.head()
```

```python
#Drop Excess columns half time
combined_epl_df = combined_epl_df.drop(columns=['Div', 'Referee', 'HS', 'AS', 'HST', 'AST', 'HF', 'AF', 'HC', 'AC', 'HY', 'AY', 'HR', 'AR', 'BWH', 'BWD', 'BWA', 'IWH', 'IWD', 'IWA', 'WHH', 'WHD', 'WHA', 'VCH', 'VCD', 'VCA'])
combined_epl_df.head()
```

```python
#Rename Columns 
combined_epl_df = combined_epl_df.rename(columns={
    "FTHG": "FT_HomeGoals", "FTAG": "FT_AwayGoals", "FTR": "FT_Result", "HTHG": "HT_HomeGoals", "HTAG": "HT_AwayGoals", "HTR": "HT_Result", "B365H" : "Bet365_HomeWin_Odds", "B365D": "Bet365_Draw_Odds", "B365A": "Bet365_AwayWin_Odds"})
combined_epl_df.head()
```

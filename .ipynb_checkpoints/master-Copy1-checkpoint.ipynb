# ML For Teams based on Home and Away Metrics

```python
#Declare Independent and Dependent Variables
lph_y_train = liverpool_home['Result']
lph_X_train = liverpool_home[['FT_HomeGoals', 'HT_HomeGoals','HomeShotsonTarget', 'HomeFouls','HomeYellowCards', 'HomeRedCards','Bet365_HomeWin_Odds', 'WilliamHill_HomeWin_Odds']]
lph_y_test = liverpool_home_t['Result']
lph_X_test = liverpool_home_t[['FT_HomeGoals', 'HT_HomeGoals','HomeShotsonTarget', 'HomeFouls','HomeYellowCards', 'HomeRedCards','Bet365_HomeWin_Odds', 'WilliamHill_HomeWin_Odds']]
```

```python
#Usage of Logistic Regression
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs',
                                max_iter=200,
                                random_state=1)
classifier
```

```python
classifier.fit(lph_X_train, lph_y_train)
```

```python
print(f"Training Data Score: {classifier.score(lph_X_train, lph_y_train)}")
print(f"Testing Data Score: {classifier.score(lph_X_test, lph_y_test)}")
```

```python
# Functions to create DataFrames for analysis 

def changeResult(s, team):
    if (s['HomeTeam'] == team) and (s['FT_Result'] == "H"):
        return 1
    elif (s['AwayTeam'] == team) and (s['FT_Result'] == "A"):
        return 1
    elif (s['HomeTeam'] == team) and (s['FT_Result'] == "D"):
        return -1
    elif (s['AwayTeam'] == team) and (s['FT_Result'] == "D"):
        return -1
    else:
        return 0
    
def Bet365(dataframe, team):
    if (dataframe['HomeTeam'] == team):
        dataframe['Bet 365 Odds'] = dataframe['Bet365_HomeWin_Odds']
        return dataframe['Bet 365 Odds']
    elif (dataframe['AwayTeam'] == team):
        dataframe['Bet 365 Odds'] = dataframe['Bet365_AwayWin_Odds']
        return dataframe['Bet 365 Odds']
    
def HomeGoals(s, team):
    if (s['HomeTeam'] == team):
        s['Goals Scored Last Game'] = s['FT_HomeGoals']
        return s['Goals Scored Last Game']
    elif (s['AwayTeam'] == team):
        s['Goals Scored Last Game'] = s['FT_AwayGoals']
        return s['Goals Scored Last Game']
        
def AwayGoals(s, team):
    if (s['HomeTeam'] == team):
        s['Goals Scored On Last Game'] = s['FT_AwayGoals']
        return s['Goals Scored On Last Game']
    elif (s['AwayTeam'] == team):
        s['Goals Scored On Last Game'] = s['FT_HomeGoals']
        return s['Goals Scored On Last Game']

def YellowCards(s, team):
    if (s['HomeTeam'] == team):
        s['Yellow Cards'] = s['HomeYellowCards']
        return s['Yellow Cards']
    elif (s['AwayTeam'] == team):
        s['Yellow Cards'] = s['AwayYellowCards']
        return s['Yellow Cards']

def YellowCardsReceived(s, team):
    if (s['HomeTeam'] == team):
        s['Yellow Cards Received'] = s['AwayYellowCards']
        return s['Yellow Cards Received']
    elif (s['AwayTeam'] == team):
        s['Yellow Cards Received'] = s['HomeYellowCards']
        return s['Yellow Cards Received']

def RedCardsReceived(s, team):
    if (s['HomeTeam'] == team):
        s['Red Cards Received'] = s['AwayRedCards']
        return s['Red Cards Received']
    elif (s['AwayTeam'] == team):
        s['Red Cards Received'] = s['HomeRedCards']
        return s['Red Cards Received']
    
def RedCards(s, team):
    if (s['HomeTeam'] == team):
        s['Red Cards'] = s['HomeRedCards']
        return s['Red Cards']
    elif (s['AwayTeam'] == team):
        s['Red Cards'] = s['AwayRedCards']
        return s['Red Cards']
    
    
def createTeamDF(dataframe, team):
    dataframe.sort_index(inplace=True)
    dataframe['Goals Scored Last Game'] = dataframe.apply(lambda x: HomeGoals(x, team), axis=1).shift(periods=1)
    dataframe['Goals Scored On Last Game'] = dataframe.apply(lambda x: AwayGoals(x, team), axis=1).shift(periods=1)
    dataframe['Yellow Cards'] = dataframe.apply(lambda x: YellowCards(x, team), axis=1).shift(periods=1)
    dataframe['Yellow Cards Received'] = dataframe.apply(lambda x: YellowCardsReceived(x, team), axis=1).shift(periods=1)
    dataframe['Red Cards'] = dataframe.apply(lambda x: RedCards(x, team), axis=1).shift(periods=1)
    dataframe['Red Cards Received'] = dataframe.apply(lambda x: RedCardsReceived(x, team), axis=1).shift(periods=1)
    dataframe['Bet365 Odds'] = dataframe.apply(lambda x: Bet365(x, team), axis=1)
    dataframe['Result'] = dataframe.apply(lambda x: changeResult(x, team), axis=1)
    dataframe.drop(columns = ['FT_HomeGoals',
                              'FT_AwayGoals',
                             'HT_HomeGoals',
                             'HT_AwayGoals',
                             'HT_Result'], inplace=True)
    dataframe.dropna(inplace=True)
    return
```

```python
createTeamDF(liverpool_df_train, 'Liverpool')
createTeamDF(liverpool_df_test, 'Liverpool')
createTeamDF(west_ham_df_train, 'West Ham')
createTeamDF(west_ham_df_test, 'West Ham')
createTeamDF(man_united_df_train, 'Man United')
createTeamDF(man_united_df_test, 'Man United')
```
